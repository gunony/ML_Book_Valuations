```python
# Select random rows
df_SelectionTemp = df_TotalBooks.sample (n=30000,random_state=1)

df_Selection = df_SelectionTemp [['book_id','title','authors','average_rating','isbn','isbn13','language_code',
                                  'num_pages','ratings_count','text_reviews_count','publication_day',
                                 'publication_month','publication_year','publisher']].copy()

# Step 1 and 2 are too long. We start by loading the df_Genre from csv file
df_Genre_Bright = pd.read_csv(r"C:/Users/gunon/Documents/bootcamp-main/3-projects/ML_Book_Valuations/Genre.csv", sep=",")

# Create a mapping dictionary containing only BookID and BookGenre
book_genre_map = dict(zip(df_Genre_Bright['bookID'], df_Genre_Bright['book_genre']))

# Convert the bookID column in dataframe Genre from object to int64
df_Genre_Bright['bookID'] = df_Genre_Bright['bookID'].astype('int64')

# Add bookgenre map to the Books dataframe using the map function
df_Selection['book_genre'] = df_Selection['book_id'].map(book_genre_map)

#Remove rows with Nan data
df_Selection = df_Selection.dropna()
```
